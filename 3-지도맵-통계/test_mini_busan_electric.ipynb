```python
# 커피 빈 데이터 메모리상에 옮기기 
import pandas as pd 
CB = pd.read_csv("./한국전력공사_부산시 전기차 충전소 충전량_20220331.csv", encoding="utf-8", index_col=0, header=0, engine="python")
CB.head()
```

```python
# 행정구역별 시군구 인구수 자료에서, -> 현재 필요한 부분은 지역 컬럼이 필요함. 정리하기. 
addr = []
for address in CB.주소:
  # 주소에 있는 값을 공백을 기준으로 나누기
  addr.append(str(address).split())
addr
```

```python
# 1차 지역명, 2차 지역명을 분리 하는 작업. 
# 1차 지역명 정리 작업 : 예) 서울, 서울시 -> 서울특별시 변경하는 작업..
addr2 =[]

for i in range(len(addr)):
  if addr[i][0] == "서울": addr[i][0] = "서울특별시"
  elif addr[i][0] == "서울시": addr[i][0] = "서울특별시"
  elif addr[i][0] == "부산시": addr[i][0] = "부산광역시"
  elif addr[i][0] == "강원도": addr[i][0] = "강원특별자치도"
  elif addr[i][0] == "인천": addr[i][0] = "인천광역시"
  elif addr[i][0] == "광주": addr[i][0] = "광주광역시"
  elif addr[i][0] == "대전시": addr[i][0] = "대전광역시"
  elif addr[i][0] == "울산시": addr[i][0] = "울산광역시"
  elif addr[i][0] == "세종시": addr[i][0] = "세종특별자치시"
  elif addr[i][0] == "경기": addr[i][0] = "경기도"
  elif addr[i][0] == "충북": addr[i][0] = "충청북도"
  elif addr[i][0] == "충남": addr[i][0] = "충청남도"
  elif addr[i][0] == "전북": addr[i][0] = "전북특별자치도"
  elif addr[i][0] == "전남": addr[i][0] = "전라남도"
  elif addr[i][0] == "경북": addr[i][0] = "경상북도"
  elif addr[i][0] == "경남": addr[i][0] = "경남북도"
  elif addr[i][0] == "제주": addr[i][0] = "제주특별자치도"
  elif addr[i][0] == "제주도": addr[i][0] = "제주특별자치도"
  elif addr[i][0] == "제주시": addr[i][0] = "제주특별자치도"

  addr2.append(" ".join(addr[i]))
addr2
```

```python
# 데이터 프레임 표 형식에 넣어서 작업 준비. 
addr2 = pd.DataFrame(addr2, columns=["address2"])

# 기존에 CB(커피빈 데이터 병합)
CB = pd.concat([CB, addr2], axis=1)
CB.head()
```

```python
# 데이터 프레임(메모리상에서 표) -> csv 파일 형태로 변환 작업. 
CB.to_csv("./CoffeeBean_2.csv", encoding="utf-8", index=False)
```

```python
import pandas as pd
import folium

# CSV 파일 로드
CB_geoData = pd.read_csv("converted_addresses.csv", encoding="utf-8-sig")

# 지도 초기화
map_CB = folium.Map(location=[35.156078, 129.059644], zoom_start=15)

# 전체 데이터 개수
total_count = len(CB_geoData)  # 전체 개수
added_count = 0    # 마커가 추가된 개수
skipped_count = 0  # NaN 값으로 인해 건너뛴 개수

# 반복문을 사용하여 마커 추가 (NaN 값이 있으면 패스)
for i, store in CB_geoData.iterrows():
    if pd.isna(store["위도"]) or pd.isna(store["경도"]):  # NaN 값이 있으면 패스
        print(f"⚠️ NaN 값이 있어 마커를 건너뜀: {store['주소']}")
        skipped_count += 1  # 건너뛴 개수 증가
        continue

    folium.Marker(
        location=[store["위도"], store["경도"]],
        popup=store["주소"],
        icon=folium.Icon(color="red", icon="star")
    ).add_to(map_CB)

    added_count += 1  # 마커 추가 개수 증가

# HTML 파일로 저장
map_CB.save("./map_부산_전기차_마커추가.html")

# 진행률 계산
added_percentage = (added_count / total_count) * 100
skipped_percentage = (skipped_count / total_count) * 100

# 최종 결과 출력
print("🎉 지도 생성 완료!")
print(f"📌 전체 데이터 개수: {total_count}")
print(f"✅ 추가된 마커 개수: {added_count} ({added_percentage:.2f}%)")
print(f"⚠️ 건너뛴 마커 개수(NaN): {skipped_count} ({skipped_percentage:.2f}%)")
print("📄 'map_부산_전기차_마커추가.html' 파일이 저장되었습니다.")
```
