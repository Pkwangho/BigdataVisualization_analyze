```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

import matplotlib.font_manager as fm
# 폰트 경로 설정
# font_path = "C:/Windows/Fonts/NanumGothic.ttf"
font_path= "C:/Windows/Fonts/malgun.ttf"  # '맑은 고딕' 폰트 경로
font_prop = fm.FontProperties(fname=font_path)

# Matplotlib 기본 폰트 설정
plt.rc('font', family=font_prop.get_name())

# 음수 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# 1. 가상의 직원 데이터셋 생성
np.random.seed(42)
employee_data = pd.DataFrame({
    'salary': np.random.normal(50000, 10000, 100),       # 평균 50,000, 표준편차 10,000
    'experience': np.random.randint(1, 21, 100),           # 1 ~ 20년의 경력
    'education': np.random.choice(['High School', 'Bachelor', 'Master', 'PhD'], 100),
    'department': np.random.choice(['Sales', 'Engineering', 'HR', 'Marketing'], 100)
})

# 2. 급여 분포 시각화 (히스토그램 + KDE)
sns.set_style("whitegrid")
plt.figure(figsize=(7,5))
sns.histplot(employee_data['salary'], kde=True, color='blue', bins=15)
plt.title("직원 급여 분포",fontproperties=font_prop)
plt.xlabel("Salary")
plt.ylabel("Count")
plt.show()

# 3. OLS 회귀 분석: Salary ~ Experience
X = employee_data[['experience']]
y = employee_data['salary']
X = sm.add_constant(X)  # 상수항 추가
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.font_manager as fm
# 폰트 경로 설정
# font_path = "C:/Windows/Fonts/NanumGothic.ttf"
font_path= "C:/Windows/Fonts/malgun.ttf"  # '맑은 고딕' 폰트 경로
font_prop = fm.FontProperties(fname=font_path)

# Matplotlib 기본 폰트 설정
plt.rc('font', family=font_prop.get_name())

# 음수 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# 1. 가상의 학생 데이터셋 생성
# 첫 번째로 np.random.seed(42)를 설정하면 난수 생성기가 고정된 시작점을 가지게 되어, np.random.rand(5) 호출 시 항상 같은 5개의 난수가 생성됩니다.
# 두 번째로 같은 시드를 다시 설정하면 동일한 순서의 난수가 생성되어, 변수 a와 b는 똑같은 값을 가지게 됩니다.
# 이렇게 시드를 고정하면 코드를 여러 번 실행해도 항상 같은 난수 배열이 생성되므로, 결과의 재현성이 보장됩니다.
np.random.seed(42)
students_data = pd.DataFrame({
    'math_score': np.random.normal(70, 10, 100),        # 평균 70, 표준편차 10
    'reading_score': np.random.normal(75, 8, 100),        # 평균 75, 표준편차 8
    'writing_score': np.random.normal(72, 9, 100),        # 평균 72, 표준편차 9
    'gender': np.random.choice(['male', 'female'], 100),
    'school_type': np.random.choice(['public', 'private'], 100)
})

# 2. 수학, 읽기, 쓰기 점수 간 상관계수 히트맵
score_cols = ['math_score', 'reading_score', 'writing_score']
corr_matrix = students_data[score_cols].corr()
plt.figure(figsize=(6,5))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("학생 점수 간 상관관계", fontproperties=font_prop)
plt.show()

# 3. 성별에 따른 수학 점수 분포 boxplot
plt.figure(figsize=(6,5))
sns.boxplot(data=students_data, x='gender', y='math_score', palette="Set2")
plt.title("성별에 따른 수학 점수 분포", fontproperties=font_prop)
plt.xlabel("Gender")
plt.ylabel("Math Score")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

import matplotlib.font_manager as fm
# 폰트 경로 설정
# font_path = "C:/Windows/Fonts/NanumGothic.ttf"
font_path= "C:/Windows/Fonts/malgun.ttf"  # '맑은 고딕' 폰트 경로
font_prop = fm.FontProperties(fname=font_path)

# Matplotlib 기본 폰트 설정
plt.rc('font', family=font_prop.get_name())

# 음수 기호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False

# 1. 가상의 전자상거래 데이터셋 생성
np.random.seed(42)
ecommerce_data = pd.DataFrame({
    'product_price': np.random.uniform(10, 100, 100),        # 10 ~ 100 사이의 제품 가격
    'units_sold': np.random.randint(50, 500, 100),             # 판매량 50 ~ 500대
    'advertisement_spend': np.random.uniform(100, 1000, 100),    # 광고비
    'discount': np.random.uniform(0, 30, 100),                 # 할인율(%)
    'rating': np.random.uniform(1, 5, 100)                     # 평점 1 ~ 5
})

# 2. 각 변수 간 관계 시각화: pairplot
# corner=True:
# pairplot의 출력에서 좌측 하단 삼각형 영역만을 표시하도록 합니다.
# 모든 변수 쌍에 대한 그래프를 그리면 상관관계 정보가 중복되므로,
# corner 옵션을 사용하면 더 간결하게 시각화할 수 있습니다.

sns.pairplot(ecommerce_data, diag_kind='kde', kind='scatter', corner=True)
plt.suptitle("전자상거래 데이터 변수 간 관계", y=1.02, fontproperties=font_prop)
plt.show()

# 3. OLS 회귀 분석: 독립 변수들이 판매량(units_sold)에 미치는 영향
X = ecommerce_data[['product_price', 'advertisement_spend', 'discount', 'rating']]
y = ecommerce_data['units_sold']
X = sm.add_constant(X)
model_ecom = sm.OLS(y, X).fit()
print(model_ecom.summary())
```
